```python
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

# 한글 폰트 설정
plt.rc('font', family='Malgun Gothic')  # Windows 사용자용
plt.rcParams['axes.unicode_minus'] = False  # 마이너스(-) 기호 깨짐 방지

# 테스트 그래프
plt.figure(figsize=(6, 3))
plt.plot([1, 2, 3], [4, 5, 6], label="테스트")
plt.xlabel("시간")
plt.ylabel("값")
plt.title("한글 폰트 테스트")
plt.legend()
plt.show()
```

```python
# 그래프 크기 설정
plt.figure(figsize=(12, 6))

# X, Y, Z 값을 시간에 따라 각각 시각화
plt.plot(df.index, df["X"], label="X 값", linestyle="-", marker="o", color="skyblue")
plt.plot(df.index, df["Y"], label="Y 값", linestyle="--", marker="*", color="pink")
plt.plot(df.index, df["Z"], label="Z 값", linestyle=":", marker="^", color="gray")

# 그래프 스타일 설정
plt.xlabel("시간") # X축 라벨
plt.ylabel("값") # Y축 라벨
plt.title("X, Y, Z 값의 시계열 변화") # 그래프 제목
plt.legend() # 범례 추가
plt.grid(True) #그리드 추가
plt.xticks(rotation=45) # 축 날짜 회전하여 가독성 증가

# 그래프 출력
plt.show()
```
